```python
import pandas as pd
```

```python
df_ivy = pd.read_csv("ivy_panda_essays.csv")

df_ivy.info()
```

```python
# func to count words
def count_words(text):
    if isinstance(text, str):
        return len(text.split())
    return 0
```

```python
# Create word count column
df_ivy["word_count"] = df_ivy["text"].apply(count_words)

# Filter rows with 200 or more words
df_ivy_length_filtered = df_ivy[df_ivy["word_count"] >= 200].copy()
```

```python
df_ivy_length_filtered.info()
```

```python
# Keep only 'text' column
df_ivy_length_filtered = (
    df_ivy_length_filtered[["text"]]
    .copy()
)

# Add new column with value 0
df_ivy_length_filtered["generated"] = 0
```

```python
df_ivy_length_filtered.head()
```

```python
df_ivy_length_filtered.head(8912).to_csv('filtered_separate/ivy_essays_8912.csv', index=False)
```

```python
# 1. List file names
files = ['filtered_separate/ivy_essays_8912.csv', 'filtered_separate/paraphrased_wiki_10k.csv', 'combined/combined_ai_human.csv']

# 2. Read and combine them into one list of DataFrames
df_list = [pd.read_csv(f) for f in files]

# 3. Concatenate them (stacking them vertically)
combined_df = pd.concat(df_list, ignore_index=True)

# 4. Save the final result
combined_df.to_csv('combined/updated_combined_ai_human.csv', index=False)

print("Files merged successfully!")
```
